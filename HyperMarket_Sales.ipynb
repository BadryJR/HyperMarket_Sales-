# Disply daily sales by day of the week

```python
dw_mapping={
    0:"Mon",
    1:"Tue",
    2:"Wed",
    3:"Thur",
    4:"Fri",
    5:"Sat",
    6:"Sun"
}
```

```python
data["Day_of_week"]=data["Date"].dt.dayofweek.map(dw_mapping)
data.head(1)
```

```python
data["Day_of_week"].value_counts().plot(kind="bar")
```

# what was the highest Month for sales ?


```python
dw_month={
    1:"Jan",
    2:"Feb",
    3:"Mar"
}
```

```python
data["Month"]=data["Date"].dt.month.map(dw_month)
data.head(1)
```

```python
data["Month"].value_counts().plot(kind="bar")
```
